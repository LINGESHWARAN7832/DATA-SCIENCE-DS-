```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="Sine Wave")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Line Plot")
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='red')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

```python
data=np.random.randn(1000)
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='blue',edgecolor='yellow', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
data=np.random.randn(1000)
df=pd.DataFrame(data, columns=['Value'])
sns.histplot(df, x='Value', kde=True)
plt.show()
```

```python
tips=sns.load_dataset('tips')
plt.figure(figsize=(6, 4))
sns.boxplot(x=tips['total_bill'])
plt.title("Box Plot of Total Bill")
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
corr_matrix = tips.select_dtypes(include=['number']).corr() # Select only numeric columns for correlation calculation.
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("correation heatmap")
plt.show()
```
